```python
from nltk.tokenize import word_tokenize 
english_punctuations = [',', '.', ':', ';', '?', '(', ')', '[', ']', '&', '!', '*', '@', '#', '$', '%','/','']  
```

```python
def create_wordlist(df=df, english_punctuations=english_punctuations):
    for each in range(len(df['Summary'])):
        word_list = [word for word in word_tokenize(df.loc[each,'Summary']) if word not in english_punctuations]
        for i in word_list:
            for j in english_punctuations:
                if j in list(i):
                    word_list += i.split(j)
                    try:
                        word_list.remove(i)
                    except:next
        word_list = [word.lower() for word in word_list if word not in english_punctuations]

        df.at[each,'Summary'] = word_list
    return df
```
